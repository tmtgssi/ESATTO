```python
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
import torch.nn as nn
from sklearn.metrics import classification_report
import joblib
from tqdm import tqdm

# Data preparation with augmentation
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Assuming that test set is in class folders
test_dataset = datasets.ImageFolder(root="./syn_vision_dataset/test", transform=transform)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# Define the feature extractor model
class FeatureExtractor(nn.Module):
    def __init__(self):
        super(FeatureExtractor, self).__init__()
        self.backbone = models.resnet152(pretrained=True)
        self.feature_extractor = nn.Sequential(*list(self.backbone.children())[:-1])  # Remove the FC layer

    def forward(self, x):
        features = self.feature_extractor(x)
        return features.view(x.size(0), -1)  # Flatten the features

# Extract features for test set
def extract_features(data_loader, model, device):
    model.eval()
    features = []
    labels = []
    with torch.no_grad():
        for images, targets in tqdm(data_loader, desc="Extracting Features"):
            images = images.to(device)
            outputs = model(images)
            features.append(outputs.cpu().numpy())
            labels.extend(targets.numpy())
    return np.vstack(features), np.array(labels)

# Initialize the feature extractor
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
feature_extractor = FeatureExtractor().to(device)

# Extract features from the test set
print("Extracting features from test set...")
test_features, test_labels = extract_features(test_loader, feature_extractor, device)

# Load the trained Random Forest classifier
rf_classifier = joblib.load("./checkpoints/ours_deep_feature_rf_classifier.pkl")

# Evaluate Random Forest on test set
print("Evaluating Random Forest classifier...")
test_predictions = rf_classifier.predict(test_features)
print(classification_report(test_labels, test_predictions, target_names=test_dataset.classes))

# Optional: Visualize some predictions
def visualize_predictions(dataset, labels, predictions, num_samples=5):
    for i in range(num_samples):
        idx = np.random.randint(0, len(dataset))  # Randomly choose an index
        image_path, _ = dataset.samples[idx]
        image = plt.imread(image_path)  # Load the image
        true_label = dataset.classes[labels[idx]]  # True label
        predicted_label = dataset.classes[predictions[idx]]  # Predicted label
        
        # Plotting the image
        plt.imshow(image)
        plt.title(f"True: {true_label}, Pred: {predicted_label}")
        plt.axis("off")
        plt.show()

# Visualize some predictions
visualize_predictions(test_dataset, test_labels, test_predictions, num_samples=5)
```
