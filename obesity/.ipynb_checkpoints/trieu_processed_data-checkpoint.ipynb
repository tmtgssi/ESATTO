```python
import timeit
import numpy as np 
import pandas as pd 
import progressbar
start = timeit.default_timer()
file_name = "./tabular_dataset/Obesity Classification.csv"
df = pd.read_csv(file_name)
print(df.isnull().values.any())
```

```python
i=0
for typ in df.Gender.unique():    
    df.loc[df["Gender"]==typ,"Gender"]=i
    print(typ, i)
    i=i+1
print(df)
```

```python
i=0
for typ in df.Label.unique():    
    df.loc[df["Label"]==typ,"Label"]=i
    print(typ, i)
    i=i+1
print(df)
```

```python
df = df.drop(columns=['ID'])
```

```python

df.to_csv('./tabular_dataset/trieu_data_processed.csv', index=False)
stop = timeit.default_timer()
print('Time: ', stop - start) 
```
