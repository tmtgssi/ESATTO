```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from torchvision import transforms, models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
# Parameters
batch_size = 1


# Data preparation with augmentation
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_set = datasets.ImageFolder(root="./syn_vision_dataset/train", transform=transform)
val_dataset = datasets.ImageFolder(root="./syn_vision_dataset/val", transform=transform)
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=False)
valid_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Define the feature extractor model
class FeatureExtractor(nn.Module):
    def __init__(self):
        super(FeatureExtractor, self).__init__()
        self.backbone = models.resnet152(pretrained=True)
        self.feature_extractor = nn.Sequential(*list(self.backbone.children())[:-1])  # Remove the FC layer

    def forward(self, x):
        features = self.feature_extractor(x)
        return features.view(x.size(0), -1)  # Flatten the features

# Extract deep features for training and validation
def extract_features(data_loader, model, device):
    model.eval()
    features = []
    labels = []
    with torch.no_grad():
        for images, targets in tqdm(data_loader, desc="Extracting Features"):
            images = images.to(device)
            outputs = model(images)
            features.append(outputs.cpu().numpy())
            labels.extend(targets.numpy())
    return np.vstack(features), np.array(labels)

# Initialize the feature extractor
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
feature_extractor = FeatureExtractor().to(device)

# Extract features for train and validation sets
print("Extracting features from train set...")
train_features, train_labels = extract_features(train_loader, feature_extractor, device)

print("Extracting features from validation set...")
val_features, val_labels = extract_features(valid_loader, feature_extractor, device)

# Train Random Forest on extracted features
print("Training Random Forest classifier...")
rf_classifier = RandomForestClassifier( n_estimators=150,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42)
rf_classifier.fit(train_features, train_labels)

# Save the trained Random Forest model
import joblib
joblib.dump(rf_classifier, "./checkpoints/ours_deep_feature_rf_classifier.pkl")

# Evaluate Random Forest
print("Evaluating Random Forest classifier...")
val_predictions = rf_classifier.predict(val_features)
print(classification_report(val_labels, val_predictions, target_names=train_set.classes))

# Optional: Visualize some predictions
def visualize_predictions(features, labels, predictions, dataset):
    for i in range(5):  # Display 5 random predictions
        idx = np.random.randint(0, len(features))
        image_path, _ = dataset.samples[idx]
        image = plt.imread(image_path)
        plt.imshow(image)
        plt.title(f"True: {dataset.classes[labels[idx]]}, Pred: {dataset.classes[predictions[idx]]}")
        plt.axis("off")
        plt.show()

visualize_predictions(val_features, val_labels, val_predictions, val_dataset)
```
