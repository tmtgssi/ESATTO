```python
import timeit
import numpy as np 
import pandas as pd 
import progressbar
start = timeit.default_timer()
file_name = "./tabular_dataset/heart disease classification dataset.csv"
df = pd.read_csv(file_name)
print(df.isnull().values.any())
df = df.dropna()
print(df.isnull().values.any())
```

```python
df.columns
```

```python
df
```

```python
i=0
for typ in df.sex.unique():    
    df.loc[df["sex"]==typ,"sex"]=i
    print(typ, i)
    i=i+1
print(df)
```

```python
i=0
for typ in df.target.unique():    
    df.loc[df["target"]==typ,"target"]=i
    print(typ, i)
    i=i+1
print(df)
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python

df.to_csv('./tabular_dataset/trieu_data_processed.csv', index=False)
stop = timeit.default_timer()
print('Time: ', stop - start) 
```
